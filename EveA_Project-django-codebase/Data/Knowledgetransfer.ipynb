```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
import requests
```

```python
data_list = []
for page in (1,2):
    url = "https://www.knowledgetransferireland.com/Events/Upcoming-Events/?pageNumber="+str(page)
    #Opening connection , grabbing the page
    uClient = uReq(url)  
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    #print(page_soup)
    div = page_soup.findAll('div',{"class":"each-item"})
    print (div)

    for container in div:
        div_tags=container.findAll('div')
        span_tags=container.div.findAll('span')
        p_tags=container.findAll('p')
        #print(p_tags[2])
        #print(div_tags[3])
        date1= span_tags[0].text
        date2= span_tags[1].text
        date3= span_tags[2].text
        title = container.h2.text
        url= container.h2.a['href']
        description=container.p.text    
        location=p_tags[2].text
        
        data = {}
        data["title"] = title
        data["url"]=url
        
        data["date"] = date1
        data["Month"] = date2
        data["Year"] = date3
        data["location"] = location
        data["Description"] = description
    
        data_list.append(data)
    
    print(data_list)
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(data_list)
df
```
