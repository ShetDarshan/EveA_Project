```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
```

```python
url ='https://www.itb.ie/NewsEvents/newsdetailmore.asp'
```

```python
#Opening connection , grabbing the page
uClient = uReq(url)
page_html = uClient.read()
```

```python
uClient.close()
```

```python
#Parsing
page_soup = soup(page_html, "lxml")
print(page_soup)
```

```python
# Finding each events
article = page_soup.findAll('div',class_='news_snippet')
print(article)
```

```python
len(article)
```

```python
data_list = []

for container in article:
# container = article[0]
    title = container.h3.text.strip()
    p_tags = container.findAll('p')
    date = p_tags[0].text.strip()
    summary = p_tags[1].text.strip()
    read_more = container.a['href']
    location = "TUD Blanchardstown"
    
    #date_time = container[0].p.strong.text
    #summary = container[1].p.strong.text
    #url = container[2].p.strong.text
    data = {}
    data["title"] = title
    data["date"] = date
    data["summary"] = summary
    data["readmore"] = read_more
    data["location"] = location
    data_list.append(data)
    
print(data_list)
```

```python
import pandas as pd
df = pd.DataFrame(data_list)
```
