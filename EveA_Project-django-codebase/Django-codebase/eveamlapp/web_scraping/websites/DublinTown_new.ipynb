```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
```

```python
data_list = []
for x in range(1,4):
    url = "https://www.dublintown.ie/whats-on/page/"+str(x)+"/"
    print(url)
    #Opening connection , grabbing the page
    uClient = uReq(url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    #print(page_soup)
    div = page_soup.findAll('article',{"class":"article whatsonarticle"})
    #print (div)

    for container in div:

        title = container.h3.text.strip()
        category=container.h5.text.strip()
        read_more= container.a['href']
        try:
            image = container.a.img['src']
        except:
            image='None'
        date = container.cite.text.strip('\n\t\t')
        a_tags=container.div.findAll('a')
    
        location=a_tags[2].text
        
        data = {}
        data["title"] = title
        data["Category"]= category
        data["image"] = image
        data["date"] = date
        data["location"] = location
        data["Read More"] = read_more
    
        data_list.append(data)
    
    print(data_list)
```

```python

import pandas as pd
df = pd.DataFrame.from_dict(data_list)
df
```
