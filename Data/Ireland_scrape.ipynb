```python
import numpy as np
import pandas as pd
import requests
import json
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import datetime
```

```python
url ='https://www.ireland.com/en-gb/what-is-available/destinations/republic-of-ireland/dublin/page/136/'
#url = 'https://www.ireland.com/en-gb/what-is-available/destinations/republic-of-ireland/dublin/all/'
#Opening connection , grabbing the page
uClient = uReq(url)
page_html = uClient.read()
uClient.close()
#Parsing
page_soup = soup(page_html, "html.parser")
print(page_soup)
```

```python
# Finding each events
article = page_soup.findAll('div',class_='content')
#print(article)
li = page_soup.findAll('ol', class_= 'listings')
print(li)
```

```python
data_list = []
for page in range(1,136):
    url ='https://www.ireland.com/en-gb/what-is-available/destinations/republic-of-ireland/dublin/page/'+str(page)+'/'
    uClient = uReq(url)
    page_html = uClient.read()
    uClient.close()
    #Parsing
    page_soup = soup(page_html, "html.parser")
    li = page_soup.findAll('ol', class_= 'listings')
    for container in li:

        title = container.h2.text
        print(title)
        location = container.h3.text
        print(location)
        description = container.p.text.strip('\n')
        print(description)
        img = container.img['src']
        print(img)
        read_more = container.a['href']
        read = 'https://www.ireland.com/'+ read_more
        print(read)
        button = page_soup.findAll('a',class_='view button')
        url = button[0]['href']
        print(url)
        
        data = {}
        data["title"] = title
        data["url"]=url
        data["read_more"] = read
        data["img"] = img
        data["location"] = location
        data["Description"] = description
    
        data_list.append(data)
    
print(data_list)
```
