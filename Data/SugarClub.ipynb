```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
import requests
```

```python
#Opening connection , grabbing the page
url="https://www.sugarclubtickets.com/"
uClient = uReq(url)
page_html = uClient.read()
uClient.close()
#Parsing
page_soup = soup(page_html, "html.parser")
#print(page_soup)
# Finding each events
div = page_soup.findAll('div',{"class":"grid-item"})
print(div)
```

```python
data_list=[]
for container in div:
    event_url=container.a['href']
    image=container.a.img['src']
    title = container.h3.text
    date=container.h2.text
    div_tags = container.findAll('div')
    time_var=date.split('/')
    time=time_var[1]
    div_tags=container.findAll('div')
    price=div_tags[1].p.text
    location="Sugar Club"
   
    
    
    data = {}
    data["Title"] = title
    data["Read More"]=event_url
    data["Time"]=time
    data["Date"]=date
    data["Price"] = price
    data["Image"] = image
    data["Venue"] = location
    
    data_list.append(data)
    
#print(data_list)
```
