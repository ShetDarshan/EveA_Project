```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
```

```python
url ='https://www.itb.ie/NewsEvents/newsdetailmore.asp'
```

```python
#Opening connection , grabbing the page
uClient = uReq(url)
page_html = uClient.read()
```

```python
uClient.close()
```

```python
#Parsing
page_soup = soup(page_html, "html.parser")
print(page_soup.prettify)
```

```python
# Finding each events
article = page_soup.findAll('div',class_='news_snippet')
print(article)
```

```python
data_list = []

for container in article:
# container = article[0]
    title = article[0].h3.text
    date_time = article[0].p.strong.text
    summary = article[1].p.strong.text
    url = article[2].p.strong.text
    data = {}
    data["title"] = title
    data["time"] = date_time
    data["summary"] = summary
    data["url"] = url
    data_list.append(data)
    
print(data_list)
```
