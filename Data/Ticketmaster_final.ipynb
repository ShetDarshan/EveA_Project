```python
import numpy as np
import pandas as pd
import requests
import json
```

```python

data_list = []
for page in range(1,102):
    
    #url = "https://app.ticketmaster.com/discovery/v2/events.json?apikey=cbpo59psc2NmPZQtJ0y7fjn65LBfF7Tz&CountryCode=IE&classificationName=Sport&page="+str(page)
    url="https://app.ticketmaster.com/discovery/v2/events.json?apikey=cbpo59psc2NmPZQtJ0y7fjn65LBfF7Tz&CountryCode=IE&city=Dublin&classificationName=Music&classificationName=Sport&classificationName=Arts,%20Theatre%20&%20Comedy,classificationName=Family%20&%20Attractions&page="+str(page)
    JSONContent=requests.get(url).json()
    content = json.dumps(JSONContent, indent = 4, sort_keys= True)
    #print(url)
    data = json.loads(content)
    
    
    var1 = data['_embedded']
    
    for var1 in var1['events']:
        data={}
        data["Event Title"]=var1["name"]
        data["Event URL"]=var1["url"]
        data["Event Date"]=var1['dates']['start']['localDate']
        try:
            data["Time"]=var1['dates']['start']['localTime']
        except:
            data["Time"]='None'
        
        try:
            data["Address Line 1"]=var1['_embedded']['venues'][0]['address']['line1']
        except:
            data["Address Line 1"]='None'
        try:
            data["Address Line 2"]=var1['_embedded']['venues'][0]['address']['line2']
        except:
            data["Address Line 2"]='None'
        try:
            data["Postal Code"]=var1['_embedded']['venues'][0]['postalCode']
        except:
            data["Postal Code"]='None'
        data["Event Image"]=var1['images'][2]['url']
        data["Event Category"]=var1['classifications'][0]['segment']['name']
        data["Event Subcategory"]=var1['classifications'][0]['genre']['name']
        data["Event Venue"]=var1['_embedded']['venues'][0]['name']

           
        #print(data)
        data_list.append(data)
        

    
       
        
```
