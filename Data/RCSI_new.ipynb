```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
import requests
```

```python
data_list = []
for page in (0,2):
    url = "https://www.rcsi.com/dublin/news-and-events/events?page="+str(page)
    #Opening connection , grabbing the page
    uClient = uReq(url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    #print(page_soup)
    div = page_soup.findAll('li',{"class":"wrapped wrapped-borderless"})
    #print (div)

    for container in div:
# container = div[0]
        title = container.h3.text.strip()
        url= container.a['href']
        #image = container.img['src']
        date = container.div.text.replace("\n"," ")
        location=container.p.text.strip('\n')
        #wh = container.p.text
        #wh = str(wh)
        #location = wh.split(')')[1].strip()

        data = {}
        data["title"] = title
        #data["image"] = image
        data["date"] = date
        data["location"] = location
        data["Read More"] = url
    
        data_list.append(data)
    
    print(data_list)
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(data_list)
df
```
